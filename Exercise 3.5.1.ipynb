```r
1.)What happens if you facet on a continuous variable? 
The continuous variable is converted to a categorical variable, and the plot contains a facet for each distinct value.
```

```r
ggplot(data = mpg) +
geom_point(mapping = aes(x = displ, y = hwy)) +
facet_wrap(~cty, nrow = 3)
```

```r
ggplot(data = mpg) +
facet_grid(drv ~ cyl)
```

```r
2.)What do the empty cells in plot with facet_grid(drv ~ cyl) mean? How do they relate to this plot?
 The empty cells (facets) in this plot are combinations of `drv` and `cyl` that have no observations.
These are the same locations in the scatter plot of `drv` and `cyl` that have no plots/data points
```

```r
3.)What plots does the following code make? What does . do?
ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = hwy)) +
  facet_grid(drv ~ .)

ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = hwy)) +
  facet_grid(. ~ cyl)
The . ignores dimension(rows or columns) when facetting.
The first code facets in rows(drv value is facetted on the y-axis) while the second code facets in columns(cyl value is facetted on the x-axis),as shown below.
```

```r
What are the advantages to using faceting instead of the colour aesthetic?
Facets allows to add more variables with aesthetics and is to split of plots into facets, subplots that each display one subset of the data.
More distinct categories can be encoded with faceting.
Easy to compare the sahpe of the relationship between x and y variables in all categories.
Reduces overlapping as the number of points or categories increase reducing the issues of color and poor visual distinction patterns forming.
What are the disadvantages?
comparison of values of categories on different plots can be cumbersome,example of 'class' variable that could be better with color aesthetic.
How might the balance change if you had a larger dataset?
as the number of categories increases, the difference between
colors decreases, to the point that the color of categories will no longer be
visually distinct.
```
